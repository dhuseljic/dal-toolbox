```python
import mlflow
import numpy as np
import pylab as plt

# client = mlflow.tracking.MlflowClient(tracking_uri='file:///mnt/work/dhuseljic/mlflow/ssal/ablation')
client = mlflow.tracking.MlflowClient(tracking_uri='file:///home/dhuseljic/mlflow/ssal/ablation')
experiment_id = '115922623489575051'
runs = client.search_runs(experiment_ids=experiment_id )
print('Found {} experiments'.format(len(runs)))
```

## Analysis of Optimizer Choice

```python
query_list = [
    "parameter.optimizer.name = 'SGD'",
    "parameter.model.scale_random_features = 'False'",
    "parameter.num_train_samples = '50'",
]
query = ' and '.join(query_list)
runs = client.search_runs(experiment_ids=experiment_id, filter_string=query)
print('Found {} experiments.'.format(len(runs)))

lr_list =[]
weight_decay_list = []
accuracy_list = []
for run in runs:
    lr = float(run.data.params['optimizer.lr'])
    weight_decay = float(run.data.params['optimizer.weight_decay'])
    accuracy = float(run.data.metrics['accuracy'])
    lr_list.append(lr)
    weight_decay_list.append(weight_decay)
    accuracy_list.append(accuracy)
```

```python
best_idx = np.argmax(accuracy_list)
print("Best lr: {:.3f}  Best weight_decay: {:.3f}".format(lr_list[best_idx], weight_decay_list[best_idx]))
plt.scatter(lr_list, weight_decay_list, c=accuracy_list, cmap='coolwarm')
plt.colorbar()
plt.xscale('log')
plt.yscale('log')
plt.show()
```
